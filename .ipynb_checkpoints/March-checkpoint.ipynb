```python
import pandas as pd
import yfinance as yf

# Input dictionary
date_ticker_dict =  {  

 '2024-03-01': ['ADANIPORTS.NS',
  'M&M.NS',
  'SBIN.NS',
  'SUNPHARMA.NS',
  'TATAMOTORS.NS']
}

# Initialize lists to store results
ticker_list = []
percentage_returns = []
first_day_closes = []
last_day_closes = []
start_dates = []  # List to store the start dates
end_dates = []    # List to store the end dates

# Define your end date
specified_end_date = "2024-03-13"

# Loop through the dictionary and download data for each date-ticker pair
for start_date, tickers in date_ticker_dict.items():
    end_date = specified_end_date  # Using the specified end date

    # Loop through tickers in this iteration
    for ticker in tickers:
        # Download historical stock data using yfinance for the specified ticker and dates
        try:
            stocks = yf.download(ticker, start=start_date, end=end_date)

            # Extracting the closing prices on the first and last day
            first_day_close = stocks['Close'].iloc[0]
            last_day_close = stocks['Close'].iloc[-1]

            # Calculating percentage return for this ticker
            percentage_return = ((last_day_close - first_day_close) / first_day_close) * 100

            # Appending the results to the lists
            ticker_list.append(ticker)
            percentage_returns.append(percentage_return)
            first_day_closes.append(first_day_close)
            last_day_closes.append(last_day_close)
            start_dates.append(pd.to_datetime(start_date))
            end_dates.append(pd.to_datetime(end_date))  # Append specified end date for each ticker
        except Exception as e:
            print(f"Error fetching data for {ticker}: {e}")

# Creating a DataFrame with ticker symbols, percentage returns, and capital information
returns_df = pd.DataFrame({
    'Start_Date': start_dates,
    'End_Date': end_dates,
    'Ticker': ticker_list,
    'Percentage_Return': percentage_returns,
    'First_Day_Close': first_day_closes,
    'Last_Day_Close': last_day_closes,
})

# Assuming an initial investment of $10,000 for each stock
initial_investment = 10000

# Calculating the capital at the last day
returns_df['Capital_At_Last_Day'] = initial_investment * (1 + returns_df['Percentage_Return'] / 100)

# Calculating the difference between capital at the last day and capital required
returns_df['Capital_Difference'] = returns_df['Capital_At_Last_Day'] - initial_investment

# Total amount invested
total_invested = initial_investment * len(returns_df)

# Total amount returned
total_returned = returns_df['Capital_At_Last_Day'].sum()

# Calculate percentage return
percentage_return = ((total_returned - total_invested) / total_invested) * 100

# Displaying the DataFrame with ticker names, percentage returns, and capital information
print(returns_df[['Start_Date', 'End_Date', 'Ticker', 'Percentage_Return', 'First_Day_Close', 'Last_Day_Close']])

# Display total amount invested and total amount returned in percentage
print("\nTotal Amount Invested: $", total_invested)
print("Total Amount Returned: $", total_returned)
print("Percentage Return: ", percentage_return, "%")

returns_df.to_csv('C:\\Users\\Mukul\\Documents\\GitHub\\CredenceLearn\\Stock_Unsupervised_baseroot\\march returns\\stock_Cluster_4_march.csv', index=False)
```
